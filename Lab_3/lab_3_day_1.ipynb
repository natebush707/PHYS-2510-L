```python
'''    Activity 1    '''
from matplotlib import pyplot as plt
import math

# build lists for given data
x = [x for x in range(1,11)]
y = [3,9,21,33,50,77,103,130,166,205]
dy = [2,4,2,4,2,4,2,4,2,4]

# build lists for ln values of given data
lnx = [math.log(num) for num in x]
lny = [math.log(num) for num in y]
lndy = [math.log(num) for num in dy]

# plotting x vs y
ax = plt.subplot()
plt.plot(x,y)
plt.xlabel('x measurements')
plt.ylabel('y measurements')
plt.title('x vs y')
plt.errorbar(x,y,yerr=dy,capsize = 10)
plt.show()

# plotting x vs ln(y), or semi-log
plt.plot(x,lny)
plt.xlabel('x measurements')
plt.ylabel('y measurements (ln)')
plt.title('x vs ln(y)')
plt.errorbar(x,lny,yerr=lndy,capsize = 10)
plt.show()

# plotting ln(x) vs ln(y), or log-log
plt.plot(lnx,lny)
plt.xlabel('x measurements (ln)')
plt.ylabel('y measurements (ln)')
plt.title('ln(x) vs ln(y)')
plt.errorbar(lnx,lny,yerr=lndy,capsize = 10)
plt.show()
```

```python
'''    Activity 2    '''
from matplotlib import pyplot as plt
import math
import pandas as pd
import numpy as np

# input raw experimental data
x = [0, 3, 6, 9, 12, 15, 18, 21, 24, 27, 30, 33, 36, 39, 42, 45, 48, 51, 54, 57, 60]
y = [7.071, 5.947, 5.005, 4.219, 3.540, 2.979, 2.505, 2.109, 1.773, 1.489, 1.250, 1.055, 0.884, 0.742, 0.625, 0.527, 0.439, 0.371, 0.313, 0.259, 0.215]
lny = [math.log(num) for num in y]

# organize data into a table for display
raw_data = {'Time(s)': x, 'Voltage(V)': y}
df = pd.DataFrame(data=raw_data)
print(df)
print()

# Graph data without linearization applied
plt.plot(x,y)
plt.xlabel('Time(s)')
plt.ylabel('Voltage(V)')
plt.title('Voltage vs Time for Discharging Capacitor')
plt.show()

# Linearize data using semi-log axes and calculate/display line of best fit
plt.plot(x,lny)
plt.xlabel('Time(s)')
plt.ylabel('Voltage(ln(V))')
plt.title('Voltage vs Time for Discharging Capacitor [semi-log]' )
model = np.polyfit(x, lny, 1)
plt.plot(x, lny, c = 'r')
plt.show()

print(f"The semi-log linear fit equation is: ln(y) = {round(model[0],3)}x + {round(model[1],2)}")
print(f"The equation for voltage over time for this discharging capacitor is: V(t) = 7e^{round(model[0],3)}t")
print(f"Tau is equal to {round(1/model[0], 2)*-1}")
```
