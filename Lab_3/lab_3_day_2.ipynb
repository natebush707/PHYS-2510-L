```python
''' Activity 3 '''
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import math

# build lists from experimental results
m_distance = [0.015, 0.02, 0.025, 0.03, 0.035, 0.04, 0.045, 0.05, 0.055, 0.06, 0.065, 0.07, 0.075, 0.08, 0.085, 0.09, 0.095, 0.1]
force_exp_1 = [3994.467, 2246.888, 1438.008, 998.617, 733.678, 561.722, 443.830, 359.502, 297.109, 249.654, 212.723, 183.419, 159.779, 140.430, 124.395, 110.957, 99.585, 89.876]

# adjust force values for linearization
ln_distance = [math.log(num) for num in m_distance]
ln_force_1 = [math.log(num) for num in force_exp_1]

# organize data into a table for display
raw_data_1 = {'Distance(m)': m_distance, 'Force(N)': force_exp_1}
df = pd.DataFrame(data=raw_data_1)
print(df)
print()

# plotting distance(m) vs force
ax = plt.subplot()
plt.plot(m_distance,force_exp_1)
plt.xlabel('Distance(cm)')
plt.ylabel('Force(N)')
plt.title('Distance vs Force for Two Equivalent Charged Spheres')
plt.show()

# plotting distance(m) vs ln(force)
ax = plt.subplot()
plt.plot(m_distance,ln_force_1)
plt.xlabel('Distance(m)')
plt.ylabel('Force(N) [log scale]')
plt.title('Distance vs Force for Two Equivalent Charged Spheres')
plt.show()

# plotting distance(m) vs ln(force)
ax = plt.subplot()
plt.plot(ln_distance,ln_force_1)
plt.xlabel('Distance(m) [log scale]')
plt.ylabel('Force(N) [log scale]')
plt.title('Distance vs Force for Two Equivalent Charged Spheres')
model = np.polyfit(ln_distance, ln_force_1, 1)
plt.plot(ln_distance, ln_force_1, c = 'r')
plt.show()

# Calculate r-squared to show how well regression line fits data
correlation_matrix = np.corrcoef(ln_distance, ln_force_1)
correlation = correlation_matrix[0,1]
r_squared = correlation**2
print(f"R squared value: {round(r_squared,2)}")

print(f"The log-log linear fit equation is: ln(y) = {round(model[0],3)}x + {round(model[1],2)}")
```
